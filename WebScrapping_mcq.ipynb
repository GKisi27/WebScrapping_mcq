```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
import csv

driver = webdriver.Chrome("F:\\Python-Jupyter-Notebook\Selenium Practice\chromedriver")#Redirect to the chromedriver in your local pc.
driver.get("https://www.mcqslearn.com/chemistry/college-chemistry-mcqs-questions-and-answers.php")#Change the url as per your requirement from the site https://www.mcqslearn.com/
driver.maximize_window()
time.sleep(1)
main_window = driver.current_window_handle
main_data = []

for link in driver.find_elements_by_xpath("/html/body/div[1]/article/ol[1]/li/div/a"):
    name = link.text
    csv_name = name.replace("/", "")
    
    ActionChains(driver) \
    .key_down(Keys.CONTROL) \
    .click(link) \
    .key_up(Keys.CONTROL) \
    .perform()

    driver.switch_to.window(driver.window_handles[1])
    time.sleep(2)
    #Write data extraction code here.
    datas = driver.find_elements_by_class_name("images_content_div")
    for data in datas:
        question = data.find_element_by_tag_name("p").text[5:]
#         print(question)
        options_data = data.find_element_by_tag_name("ol")
        options = options_data.find_elements_by_tag_name("li")
        options_list = []
        for option in options:
            options_list.append(option.text)
        ans = data.find_element_by_class_name("phover")
        ans1 = ans.find_element_by_tag_name("button").click()
        correct_ans = ans.find_element_by_tag_name("span").text
#         print(correct_ans)
        
        temp_dict = {}
        temp_dict["QUESTION"] = question
        temp_dict["OPTION A"] = options_list[0]
        temp_dict["OPTION B"] = options_list[1]
        temp_dict["OPTION C"] = options_list[2]
        temp_dict["OPTION D"] = options_list[3]
        temp_dict["CORRECT ANSWER"] = correct_ans
        main_data.append(temp_dict)
#     print(main_data)
#     Write code here to write data to csv file
    columns = [key for key in main_data[0]]
    with open(f"{csv_name}.csv", "w", encoding="utf-8", newline = "") as wf:
        writer = csv.DictWriter(wf, fieldnames = columns)
        writer.writeheader()
        writer.writerows(main_data)
    main_data.clear()
    driver.close()
    driver.switch_to.window(main_window)

time.sleep(2)
# Close the instance of the browser.
driver.quit()
```
